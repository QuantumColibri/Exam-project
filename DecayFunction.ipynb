```python
import numpy as np
```

```python
def decay_two_identical(parent_mass, daugther_mass):
    if parent_mass < 2 * daugther_mass:
        raise ValueError('Error! Decay kinematically forbidden M < 2*m')
    E_daughter = parent_mass/2
    p_daughter = 0.5 * np.sqrt(parent_mass**2 - 4* daugther_mass**2)
    
    theta = np.random.uniform(0, 2*np.pi)
    phi = np.random.uniform(0, 2*np.pi)

    px = p_daughter * np.cos(theta) * np.cos(phi)
    py = p_daughter * np.cos(theta) * np.sin(phi)
    pz = p_daughter * np.sin(theta)

    p1 = np.array([px, py, pz])
    p2 = -p1

    return (E_daughter, p1), (E_daughter, p2)
```

```python
decay_two_identical(122, 45)
```
